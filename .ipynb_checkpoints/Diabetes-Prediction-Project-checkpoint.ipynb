This project seeks to build a **robust machine learning model** to predict diabetes risk, enabling **early intervention** and improved patient care.

```python
# Import Dependencies

import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn import svm
import seaborn as sns
from sklearn.metrics import accuracy_score
import matplotlib
import matplotlib.pyplot as plt
import plotly.express as px

path = "diabetes_prediction_dataset.csv"
diabetes_data = pd.read_csv(path)

diabetes_data.sample(4)

diabetes_data.shape

# Displaying basic info
print("Dataset Info:")
print(diabetes_data.info())

# Checking for missing values
print("\nMissing Values:")
print(diabetes_data.isnull().sum())

# Checking for duplicate rows
print("\nDuplicate Rows:", diabetes_data.duplicated().sum())

# Removing duplicate rows
df_cleaned = diabetes_data.drop_duplicates()

# Verifying that duplicates are removed
print("Duplicate Rows After Cleaning:", df_cleaned.duplicated().sum())

# Running the Summary statistics to check for any inconsistencies
print("\nSummary Statistics:")
df_cleaned.describe()

# Removing away rows where age is less than 1
df_cleaned = df_cleaned[df_cleaned['age'] >= 1]

# Verifying the change
print("Minimum Age After Cleaning:", df_cleaned['age'].min())
print("Summary Statistics After Age Cleaning:")
print(df_cleaned['age'].describe())

# Checking for unique the values in categorical columns
categorical_cols = ['gender', 'hypertension', 'heart_disease', 'smoking_history', 'diabetes']
for col in categorical_cols:
    print(f"\nUnique values in {col}: {df_cleaned[col].unique()}")

# Handling "Other" in gender (droping them)
df_cleaned = df_cleaned[df_cleaned["gender"] != "Other"]  # If you want to remove

df_cleaned.shape

df_cleaned["smoking_history"] = df_cleaned["smoking_history"].replace({
    "not current": "former",  # This clearly means they quit
    "No Info": "unknown" 
})

# Checking for outliers using boxplot (requires matplotlib & seaborn)
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10,6))
sns.boxplot(data=df_cleaned[['age', 'bmi', 'HbA1c_level', 'blood_glucose_level']])
plt.title("Boxplot of Numeric Features (Outlier Detection)")
plt.show()

plt.figure(figsize=(12,6))
# to slant the country in x axis
plt.xticks(rotation=75)
plt.title("Hypertension vs Diabetes")
sns.barplot(x=df_cleaned.hypertension, y=df_cleaned.diabetes);

plt.figure(figsize=(12,6))
# to slant the country in x axis
plt.xticks(rotation=75)
plt.title("Heart disease vs Diabetes")
sns.barplot(x=df_cleaned.heart_disease, y=df_cleaned.diabetes);

df_cleaned.corr(numeric_only=True)

sns.heatmap(df_cleaned.corr(numeric_only=True), cmap='Reds', annot=True)
plt.title('Correlation Matrix');

labels = df_cleaned.drop(columns = 'diabetes',axis = 1)
target = df_cleaned['diabetes']

labels.head(2)

target

# One-Hot Encoding
labels = pd.get_dummies(labels, drop_first=True)  # Convert categorical columns to numbers

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(labels)

standardized_data = scaler.transform(labels)

standardized_data

labels = standardized_data
target=df_cleaned['diabetes']

print(labels)
print(target)

labels_train,labels_test,target_train,target_test = train_test_split(labels,target,test_size = 0.2,stratify = target,random_state=0)

print(labels.shape,labels_train.shape,labels_test.shape)

classifier = svm.SVC(kernel='linear') 

# training the support vector machine classifier

from sklearn.linear_model import LogisticRegression

classifier = LogisticRegression(class_weight='balanced')
classifier.fit(labels_train, target_train)

# Finding accuracy score on the training data
labels_train_prediction = classifier.predict(labels_train)
training_data_accuracy = accuracy_score(labels_train_prediction,target_train)

print("The accuracy score of the training data: ", training_data_accuracy)

labels_test_prediction = classifier.predict(labels_test)
test_data_accuracy = accuracy_score(labels_test_prediction, target_test)
print("The accuracy score of the test data: ", test_data_accuracy)

from sklearn.metrics import classification_report
print(classification_report(target_test, labels_test_prediction))

import numpy as np

def predict_diabetes():
    print("Enter the following health details:")
    
    # Collect user input
    age = float(input("Age: "))
    hypertension = int(input("Hypertension (0: No, 1: Yes): "))
    heart_disease = int(input("Heart Disease (0: No, 1: Yes): "))
    bmi = float(input("BMI: "))
    HbA1c_level = float(input("HbA1c Level: "))
    blood_glucose_level = float(input("Blood Glucose Level: "))

    # Store inputs as a NumPy array (reshaped for model)
    user_data = np.array([[age, hypertension, heart_disease, bmi, HbA1c_level, blood_glucose_level]])

    # Make prediction
    prediction = classifier.predict(user_data)

    # Output result
    if prediction[0] == 1:
        print("\n🔴 High Risk of Diabetes!")
    else:
        print("\n🟢 Low Risk of Diabetes!")

# Run the function
predict_diabetes()




# Encoding categorical variables
from sklearn.preprocessing import LabelEncoder

# Convert 'gender' and 'smoking_history' to numeric labels
encoder = LabelEncoder()
df_cleaned['gender'] = encoder.fit_transform(df_cleaned['gender'])
df_cleaned['smoking_history'] = encoder.fit_transform(df_cleaned['smoking_history'])



import numpy as np

def predict_diabetes():
    print("Enter the following health details:")
    
    # Collect user input
    age = float(input("Age: "))
    hypertension = int(input("Hypertension (0: No, 1: Yes): "))
    heart_disease = int(input("Heart Disease (0: No, 1: Yes): "))
    bmi = float(input("BMI: "))
    HbA1c_level = float(input("HbA1c Level: "))
    blood_glucose_level = float(input("Blood Glucose Level: "))

    # Store inputs as a NumPy array (reshaped for model)
    user_data = np.array([[age, hypertension, heart_disease, bmi, HbA1c_level, blood_glucose_level]])

    # Make prediction
    prediction = classifier.predict(user_data)

    # Output result
    if prediction[0] == 1:
        print("\n🔴 High Risk of Diabetes!")
    else:
        print("\n🟢 Low Risk of Diabetes!")

# Run the function to test prediction
# predict_diabetes()  # Uncomment to allow user input
```

# Predicting System

This is going to predict based on 0 and 1

- 1 > Diabetic

```python
import numpy as np

def predict_diabetes():
    print("Enter the following health details:")
    
    # Collect user input
    age = float(input("Age: "))
    hypertension = int(input("Hypertension (0: No, 1: Yes): "))
    heart_disease = int(input("Heart Disease (0: No, 1: Yes): "))
    bmi = float(input("BMI: "))
    HbA1c_level = float(input("HbA1c Level: "))
    blood_glucose_level = float(input("Blood Glucose Level: "))

    # Store inputs as a NumPy array (reshaped for model)
    user_data = np.array([[age, hypertension, heart_disease, bmi, HbA1c_level, blood_glucose_level]])

    # Make prediction
    prediction = classifier.predict(user_data)

    # Output result
    if prediction[0] == 1:
        print("\n🔴 High Risk of Diabetes!")
    else:
        print("\n🟢 Low Risk of Diabetes!")

# Run the function
predict_diabetes()
```
